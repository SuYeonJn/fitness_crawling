```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import xlsxwriter

workbook = xlsxwriter.Workbook('fitness3.xlsx')
worksheet = workbook.add_worksheet()

driver = webdriver.Chrome(service= Service(ChromeDriverManager().install()))

url = "https://m.map.naver.com/search2/search.naver?query=%EA%B4%80%EC%95%85%EA%B5%AC%20%ED%97%AC%EC%8A%A4%EC%9E%A5&sm=hty&style=v5"
driver.get(url)
time.sleep(2)
html = driver.page_source

soup = BeautifulSoup(html, 'html.parser')

wrap = soup.select('li._item._lazyImgContainer')

"""
title_list = []
section_list = []
address_list = []
stars_list = []
price_list=[]
site_list=[]
"""

row = 0

btn = driver.find_elements(By.CLASS_NAME, 'item_tit._title')
for btns in btn:
    btns.click()
    time.sleep(1.5)
    
    title= driver.find_elements(By.CLASS_NAME, 'Fc1rA')[0].text
    section= driver.find_elements(By.CLASS_NAME, 'DJJvD')[0].text
    address= driver.find_elements(By.CLASS_NAME, 'IH7VW')[0].text
    try:
        stars= driver.find_elements(By.CLASS_NAME, 'PXMot.LXIwF')[0].text.lstrip('별점\n')
    except:
        stars = ''
    
    try:
        price = driver.find_elements(By.CLASS_NAME, 'X7q00.WVr9W')[0].text.strip()
    except:
        price = ''
        
    try:
        site = driver.find_elements(By.CLASS_NAME, 'CQDdi')[0].text.strip()
    except:
        site = ''
    
    try:
        review = driver.find_elements(By.CLASS_NAME, 'place_bluelink')[0].text
    except:
        review =''
        
    url = driver.current_url
    
    time.sleep(1)

    try: 
        picture = driver.find_elements(By.CLASS_NAME, 'xez5V')[0]
        picture.click()
        time.sleep(1)
        picture1 = driver.find_elements(By.CLASS_NAME, 'wzrbN')[0]
        image=  picture1.find_element(By.TAG_NAME, 'img')
        thumbnail = image.get_attribute("src")
    except:
        picture2 = driver.find_elements(By.CLASS_NAME, 'CEX4u')[0]
        picture2.click()
        time.sleep(1)
        picture1 = driver.find_elements(By.CLASS_NAME, 'wzrbN')[0]
        image=  picture1.find_element(By.TAG_NAME, 'img')
        thumbnail = image.get_attribute("src")
        bar = driver.find_elements(By.CLASS_NAME, 'flicking-viewport')[0]
        home = bar.find_elements(By.TAG_NAME,'a')[0]
        home.click()
        time.sleep(1)
   
    
    """
    title_list.append(title)
    section_list.append(section)
    address_list.append(address)
    stars_list.append(stars)
    price_list.append(price)
    site_list.append(site)
    """
    
    time.sleep(0.5)
    
    print(title, section, address, stars, price, site, review, url, thumbnail)
    

    worksheet.write(row, 0, title)
    worksheet.write(row, 1, section)
    worksheet.write(row, 2, address)
    worksheet.write(row, 3, stars)
    worksheet.write(row, 4, price)
    worksheet.write(row, 5, site)
    worksheet.write(row, 6, review)
    worksheet.write(row, 7, url)
    worksheet.write(row, 8, thumbnail)
    
    row += 1
    
    back = driver.find_elements(By.CLASS_NAME, 'DDfpb')[0]
    back.click()
    time.sleep(1)
    back.click()
    time.sleep(1)
workbook.close()
```
